```python
# 正确使用通义千问的导入方式
from langchain_community.llms import Tongyi  # 注意使用Tongyi而非OpenAI

llm = Tongyi(
    model_name="qwen-plus",  # 支持 qwen-turbo/qwen-plus/qwen-max
    temperature=0.9,
    max_tokens=200
)

text = llm("请给我写一句情人节红玫瑰的中文宣传语，不少于30个字")
print(text)
```
